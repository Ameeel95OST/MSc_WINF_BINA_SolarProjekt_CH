```python
import pandas as pd  # Importing pandas for data manipulation

# Load CSV File
file_path = 'data/electricity_production_plants/000_ElectricityProductionPlant.csv'  # Replace with the path to your CSV file
df = pd.read_csv(file_path, sep=';')  # Read the CSV file into a DataFrame, handling bad lines and ensuring comma as separator

# Wertetabelle MainCategory laden
main_categories_file_path = 'data/electricity_production_plants/MainCategoryCatalogue.csv'  # Path to the replacement file
main_categories = pd.read_csv(main_categories_file_path, sep=';')

# Annahme: Die Wertetabelle hat zwei Spalten: 'old_value' und 'new_value'
# Erstellen eines Mappings aus der Wertetabelle
main_categories_dict = dict(zip(main_categories['Catalogue_id'], main_categories['de']))

# Ersetzen der Wwerte in der DataFrame-Spalte 'MainCategory' mit den neuen Werten aus der Wertetabelle
df['MainCategory'] = df['MainCategory'].replace(main_categories_dict)

# Display the first few rows of the DataFrame to verify the data
df.head()
```
